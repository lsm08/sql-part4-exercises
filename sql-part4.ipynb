1. Write a query that will return the number of users who rated a book above it's average rating.

```sql
use BOOKSDB;

select top 2 * from dbo.books;
select top 2 * from BooksDB.dbo.ratings;


select count(*) from dbo.ratings as r where r.rating > ( select b.average_rating
from dbo.books as b where r.book_id = b.book_id);
```
